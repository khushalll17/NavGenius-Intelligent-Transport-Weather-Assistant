```python
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-3.3-70B-Instruct")
model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-3.3-70B-Instruct")
```

```python
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
prompt = PromptTemplate(
    input_variables=["origin", "destination"],
    template="""You are an AI assistant. Your task is to help the user determine the best mode of transport from the origin to the destination, considering the following factors:

    1. **Origin**: {origin}
    2. **Destination**: {destination}

    Based on the data you have access to (e.g., transportation modes, weather information, and real-time data), please provide the following:

    1. **Modes of Transport**: List all possible modes of transport available from the origin to the destination (e.g., car, bus, subway, flight, etc.). For each mode, include the following:
       - Estimated **cost** of travel in Indian Rupees (₹)
       - Estimated **time** it will take to reach the destination
       - **Next available transport times** for each mode of transport over the next 24 hours
       - **Stations or pickup points** where this mode of transport is available
       - Whether **weather conditions** will affect the transport, and in what way (e.g., rain causing delays in public transport, road closures due to snow, etc.)
       - If there is any **delay** expected due to weather, provide an estimate of how much extra time it will take to reach the destination.

    2. **Weather Conditions**: Retrieve the current weather information for both the origin and destination. Include:
        - Current **weather** at both locations (e.g., sunny, rainy, snowing, etc.)
        - **Forecast for the next 24 hours** for both locations
        - How these weather conditions may **affect the travel time** or mode of transport (e.g., traffic due to rain, delays on flights due to snow, etc.)

    3. **Estimated Delay**: Based on the weather and real-time traffic data, estimate how much **delay** the user should expect to encounter for each mode of transport. Include any known traffic conditions, road closures, or public transport delays.

    4. **User Selection**: Allow the user to **select a mode of transport** from the provided options.
       - If the user selects **Car**, generate:
         - The **best route** to travel between the origin and destination.
         - Provide a **Google Maps route link**.
         - Show alternative routes if available.

    5. **Summary**: After considering the above factors, provide a **recommendation** for the best mode of transport based on:
       - Cost-effectiveness
       - Time efficiency
       - Comfort (if relevant to the user's query)
       - Weather conditions and potential delays

    Ensure that all details are clearly organized and include real-time data wherever possible.
"""
)
```

```python
from langchain.schema.runnable import RunnableLambda
```

```python
def generate_text(prompt_value, **kwargs):
    prompt_text = str(prompt_value)
    response = pipe(prompt_text, max_length=500, truncation=True)
    return response[0]["generated_text"]
```

```python
llm = RunnableLambda(generate_text)
```

```python
transport_chain = prompt | llm
```

```python
transport_tool = Tool(
    name="Transport Guide",
    func=transport_chain.invoke,
    description="Provides the best transport options between an origin and a destination, considering cost, time, weather, and delays."
)
```

```python
tools = load_tools(["serpapi"])
tools.append(transport_tool)
```

```python
agent = initialize_agent(
    tools, llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)
```

```python
response = agent.invoke("What is the best way to travel from Delhi to Mumbai?")
print(response)
```
